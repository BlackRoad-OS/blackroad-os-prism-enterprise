# Demo C — AM-4 Energy Ledger
Track the energy increment versus phase displacement at fixed entropy.

```python
import numpy as np
import matplotlib.pyplot as plt
from amundson_blackroad.thermo import spiral_entropy, energy_increment, landauer_min

T = 300.0
theta = np.linspace(0.5, 2.0, 50)
a = 0.3
S = [spiral_entropy(a, th) for th in theta]
delta_a = 0.01
delta_theta = float(theta[1] - theta[0])
delta_E = [energy_increment(T, delta_a, delta_theta, a, th, Omega=1e-21) for th in theta]
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(theta, delta_E)
ax.set_xlabel('theta [rad]')
ax.set_ylabel('ΔE [J]')
ax.set_title('Energy increment vs phase')
fig.tight_layout()
landauer_min(T, 1.0)
```
