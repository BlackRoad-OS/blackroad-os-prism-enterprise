# Demo A — AM-2 Phase Portrait
Simulate coupled amplitude–phase dynamics and visualise stability.

```python
import numpy as np
import matplotlib.pyplot as plt
from amundson_blackroad.spiral import simulate_am2, fixed_point_stability

t, a, theta, amp = simulate_am2(T=5.0, dt=1e-3, a0=0.2, theta0=0.0, gamma=0.3, kappa=0.6, eta=0.4)
report = fixed_point_stability(a[-1], theta[-1], gamma=0.3, kappa=0.6, eta=0.4, omega0=1.0)
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(theta, a)
ax.set_xlabel('theta [rad]')
ax.set_ylabel('a [1]')
ax.set_title('AM-2 Phase Portrait')
fig.tight_layout()
report.eigenvalues
```
