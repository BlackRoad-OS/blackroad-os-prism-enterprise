# Demo B — BR-1/2 Autonomy Transport
Evolve an autonomy density across a dual-well trust potential.

```python
import numpy as np
import matplotlib.pyplot as plt
from amundson_blackroad.autonomy import step_transport, conserved_mass

x = np.linspace(-5.0, 5.0, 512)
dx = x[1] - x[0]
A = np.exp(-x**2)
rho = -(np.exp(-(x - 1.5)**2) + np.exp(-(x + 1.5)**2))
trajectory = [A.copy()]
for _ in range(400):
    A, _ = step_transport(A, rho, dx, dt=1e-3, mu_A=1.0)
    trajectory.append(A.copy())
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(x, trajectory[0], label='initial')
ax.plot(x, trajectory[-1], label='final')
ax.set_xlabel('x [m]')
ax.set_ylabel('A [A-units]')
ax.legend()
fig.tight_layout()
conserved_mass(x, trajectory[-1])
```
